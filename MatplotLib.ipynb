# ðŸ“Š Barplot

### plt.bar(x, y, color=" ", edgecolor=" ", linewidth= , label=" ")

```python
x = ["A","B","C","D","E"]
y = [1000,1500,1700,1600,2200]
plt.bar(x, y, color="tomato", edgecolor="black", linewidth=1.5, label="Sales 2025")
plt.title("Product vs Sales")
plt.xlabel("Product Category")
plt.ylabel("Sales in Rs")
plt.legend()
plt.show()
```

# ðŸ“Š Pie Chart

### plt.pie(x, labels=y, colors = [ ], wedgeprops={"edgecolor": " ", "linewidth": 0.5} , autopct="%1.1f%%")

```python
students = [600, 150, 100, 200, 500] # Numerical Data
languages = ["Python", "Java", "C++", "JS", "C"] # Categorical Data
plt.pie(students, labels=languages, colors = ["pink","lawngreen","lightskyblue","khaki","tomato"],
        wedgeprops={"edgecolor": "black","linewidth": 0.5, "linestyle": "--"} , autopct="%1.2f%%")
plt.title("Coding Language Popularity")
plt.show()
```

```python
# plt.subplot(nrow, ncol, index)
```

```python
# 2 rows, 3 columns â†’ total 6 possible subplots

plt.subplot(2, 3, 1)   # Row 1, Col 1
plt.plot([1, 2, 3], [1, 2, 3])
plt.title("1")
plt.xlabel("x1")
plt.ylabel("y1")

plt.subplot(2, 3, 2)   # Row 1, Col 2
plt.plot([1, 2, 3], [1, 2, 1.5])
plt.title("2")
plt.xlabel("x2")
plt.ylabel("y2")

plt.subplot(2, 3, 3)   # Row 1, Col 3
plt.plot([1, 2, 3], [1, 8, 27])
plt.title("3")
plt.xlabel("x3")
plt.ylabel("y3")

plt.subplot(2, 3, 4)   # Row 2, Col 1
plt.plot([1, 2, 3], [1, 2, 2.5])
plt.title("4")
plt.xlabel("x4")
plt.ylabel("y4")

plt.subplot(2, 3, 5)   # Row 2, Col 2
plt.plot([1, 2, 3], [1, 32, 243])
plt.title("5")
plt.xlabel("x5")
plt.ylabel("y5")

plt.subplot(2, 3, 6)   # Row 2, Col 3
plt.plot([1, 2, 3], [5, 1, 5])
plt.title("6")
plt.xlabel("x6")
plt.ylabel("y6")

plt.tight_layout()
plt.show()
```

```python
# fig, ax = plt.subplots(nrows, ncols, figsize=(width, height))
# Figure (fig) = the entire canvas
# Axes (ax) = small rectangles on that paper where each graph is drawn

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

x = [1,2,3,4]
y = [10,20,15,25]

ax[0].plot(x,y,linestyle="--",color="deeppink")
ax[0].set_title("Line Chart")
ax[0].set_xlabel("X-axis")
ax[0].set_ylabel("Y-axis")

ax[1].bar(x,y,color="dodgerblue",edgecolor="black",linewidth=1)
ax[1].set_title("Bar Chart")
ax[1].set_xlabel("X-axis")
ax[1].set_ylabel("Y-axis")

fig.suptitle("Comparison")

plt.tight_layout() # always before plt.show()
plt.show()
```
